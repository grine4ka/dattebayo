```python
import matplotlib.pyplot as plt

from mplsoccer import Pitch, VerticalPitch
```

```python
pitch = Pitch(pitch_color='grass', line_color='white', stripe=True)
fig, ax = pitch.draw()
```

```python
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(10, 6), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')
```

```python
pitch = VerticalPitch()
fig, ax = pitch.draw()
```

```python
pitch = Pitch()
fig, axs = pitch.grid(nrows = 3, ncols = 2)
```

```python
pitch = Pitch(half=True)

fig, ax = pitch.draw()
```

```python
pitch = VerticalPitch(half=True, pitch_type='opta', pitch_color='grass', line_color='white', stripe=True)
fig, ax = pitch.draw()
```
