# Statsbomb API

Example of how to work with [statsbombpy](https://github.com/statsbomb/statsbombpy) based on [video](https://www.youtube.com/watch?v=Ryn8etss1p4) of [McKay Johns](https://www.youtube.com/channel/UCmqincDKps3syxvD4hbODSg)

```python
from statsbombpy import sb
```

```python
sb.competitions()
```

```python
sb.matches(competition_id = 43, season_id = 3)
```

```python
events = sb.events(match_id = 8657)
```

```python
events.columns
```

```python
events = events[['team','type','minute','location','pass_end_location','player']]
events = events[events['team'] == 'England'].reset_index()
```
